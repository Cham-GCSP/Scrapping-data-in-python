This code is to scrap list of dataset name from OGD
After that, randomly select data list from OGD and KOSIS

```python
import os

from selenium import webdriver

from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup

import pandas as pd

import csv

import random
```

```python
from platform import python_version
print(python_version())
```

```python
# scaping data from English website of OGD

driver = webdriver.Chrome()

title = []

for p in range(96):
    
    current_page = p+1
    
    val = 'https://www.data.go.kr/en/tcs/dss/selectDataSetList.do?dType=FILE&keyword=&detailKeyword=&publicDataPk=&recmSe=&detailText=&relatedKeyword=&commaNotInData=&commaAndData=&commaOrData=&must_not=&tabId=&dataSetCoreTf=&coreDataNm=&sort=&relRadio=&orgFullName=&orgFilter=&org=&orgSearch=&currentPage={}&perPage=10&brm=Health+care&instt=Administrative+Agency%2Cpublic+institutions&svcType=&kwrdArray=&extsn=&coreDataNmArray=&pblonsipScopeCode='.format(current_page)
    
    wait = WebDriverWait(driver, 10)

    driver.get(val)

    get_url = driver.current_url

    wait.until(EC.url_to_be(val))
    

    if get_url == val:
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        data = soup.find(class_= 'result-list')
        for t in data.find_all(class_='title'):
            title.append(t.text.strip())
        final = pd.DataFrame(title)
        
final.to_csv('ogd_en_1409.csv', encoding = 'utf-8')
```

```python
# scaping data from Korean website of OGD

driver = webdriver.Chrome()

title1 = []
for p in range(95):
    
    current_page = p+1
    
    val = 'https://www.data.go.kr/tcs/dss/selectDataSetList.do?dType=FILE&keyword=&detailKeyword=&publicDataPk=&recmSe=&detailText=&relatedKeyword=&commaNotInData=&commaAndData=&commaOrData=&must_not=&tabId=&dataSetCoreTf=&coreDataNm=&sort=&relRadio=&orgFullName=&orgFilter=&org=&orgSearch=&currentPage={}&perPage=10&brm=%EB%B3%B4%EA%B1%B4%EC%9D%98%EB%A3%8C&instt=%EA%B3%B5%EA%B3%B5%EA%B8%B0%EA%B4%80%2C%EA%B5%AD%EA%B0%80%ED%96%89%EC%A0%95%EA%B8%B0%EA%B4%80%2C%ED%97%8C%EB%B2%95%EA%B8%B0%EA%B4%80&svcType=&kwrdArray=&extsn=&coreDataNmArray=&pblonsipScopeCode='.format(current_page)
    
    wait = WebDriverWait(driver, 10)

    driver.get(val)

    get_url = driver.current_url

    wait.until(EC.url_to_be(val))
    

    if get_url == val:
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser',fromEncoding='cp949')
        data = soup.find(class_= 'result-list')
        for t in data.find_all(class_='title'):
            title1.append(t.text.strip())
        final1 = pd.DataFrame(title1)

final1.to_csv('ogd_kr_1409.csv', encoding = 'cp949')
```

Randomly select datasets

```python
# OGD list
final_ogd_en = random.sample(title, k = 50)
final_ogd_kr = random.sample(title1, k = 50)

final_data_ogd_en = pd.DataFrame(final_ogd_en)
final_data_ogd_kr = pd.DataFrame(final_ogd_kr)

final_data_ogd_en.to_csv('final_ogd_en.csv', encoding = 'cp949')
final_data_ogd_kr.to_csv('final_ogd_kr.csv', encoding = 'cp949')
```
